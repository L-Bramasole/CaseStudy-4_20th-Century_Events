## Importing libraries and data

```python
import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis
```

```python
# Import relationship data

relationship_df = pd.read_csv('20th_century_countries_relationship.csv', index_col = 0)
```

```python
relationship_df
```

## Graph visualization - Panda

```python
# Create a graph from a pandas dataframe

G = nx.from_pandas_edgelist(relationship_df, 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

```python
h = plt.figure(figsize=(7,7))
pos = nx.kamada_kawai_layout(G)

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_color='Maroon')
nx.draw_networkx_edges(G, pos, edge_cmap=plt.cm.Blues)

# Draw labels with custom font and offset
nx.draw_networkx_labels(
    G,
    pos,
    font_family='Arial',
    font_size=12,
    verticalalignment='bottom',   # puts text above the node
)

plt.axis('off')
plt.show()
```

```python
h.savefig('networkx_2othCentury.png')
```

## Community detection - Leiden algorithm

```python
# First, install the required dependencies
import sys
!{sys.executable} -m pip install --upgrade python-igraph leidenalg

```

```python
# Then import and use the leiden algorithm
import igraph
import leidenalg
from cdlib import algorithms

coms = algorithms.leiden(G)
```

```python
type(coms)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k,v in new_com.items()}
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width="1300px", height="700px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_ = True)
com_net.repulsion()
com_net.show("countries_communities_leiden.html")
```

## The most important countries in the 20th century events 

## Degree centrality

```python
degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
degree_df = pd.DataFrame(degree_dict.items(), columns=['countries','centrality'])
```

```python
degree_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
# Plot the degree centrality

plt.figure(figsize = (8, 8))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = degree_df).set_title("Countries in 20th Century- degree centrality")
```

## Closeness centrality


```python

closeness_dict = nx.closeness_centrality(G)
closeness_df = pd.DataFrame(closeness_dict.items(), columns=['countries','centrality'])
```

```python
closeness_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (10, 8))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = closeness_df).set_title("Countries in 20th Century- closeness centrality")
```

## Betweenness centrality

```python
betweenness_dict = nx.betweenness_centrality(G)
betweennes_df = pd.DataFrame(betweenness_dict.items(), columns=['countries','centrality'])
```

```python
betweennes_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (10, 8))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = betweennes_df).set_title("Countries in 20th Century- betweenness centrality")
```

```python
# Save centrality measures
nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```

# Network Analysis results: 20th-Century Geopolitical Structures

This report identifies the structural anchors of the 20th-century geopolitical network using the Lyndon community algorithm and three centrality metrics. The data reveals a tri-polar core (Germany, Soviet Union, and Japan) that controls both direct influence and global information flow.

Primary Cluster: A high-density community exists between Germany, Italy, Japan, and Poland.
Directional Influence: France exerts a significant "thick" influence on Germany.
Core Hubs: Germany, Japan, and France serve as the primary functional centers.

Centrality Results

Metric;	          Leader(s)	  ;                                  Role in Network

Centrality   ;	Germany, Japan, Soviet Union             ;     	Most direct allies and interactions.

Closeness	 ;   Soviet Union, Germany, Japan, Poland, France;	Most efficient at spreading influence.

Betweenness	  ;  Soviet Union	                           ;     The ultimate bridge/gatekeeper of information.

These results are structurally sound and geopolitically accurate:

Strategic Positioning: The Soviet Union leading in Betweenness and Closeness matches its historical role as a bridge between eastern and western blocs.

Cluster Accuracy: The Lyndon algorithm correctly grouped the historically aligned Axis powers.

Hierarchy: Chinas peripheral status (lowest in all metrics) correctly places it as an emergent node rather than a central 20th-century gatekeeper.
