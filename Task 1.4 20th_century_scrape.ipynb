# Data Scraping from Wikipedia

```python
# Scraping from url
driver.get("https://en.wikipedia.org/wiki/Key_events_of_the_20th_century")

```

```python
print(driver.current_url)
```

```python
import time
time.sleep(2)
```

```python
# Get HTML
html = driver.page_source
driver.quit()
```

```python
# Parse HTML. the soup makes the html bettre to read and more structured
soup = BeautifulSoup(html, "html.parser")
```

```python
text = soup.get_text(separator="\n")
#The separator="\n" just makes the text easier to read later.
```

```python
# cleaning the text will remove: empty lines, weird spacing, stray tabs
clean_text = "\n".join([line.strip() for line in text.splitlines() if line.strip()])
```

```python
# save the scraped text to txt file
with open("key_events_20th_century.txt", "w", encoding="utf-8") as f:
    f.write(clean_text)
```

# Scraping Countries list for task 1.5

```python
# Initialize the browser with the new selenium. driver2 name baecause we used driver name before
driver2 = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# Scraping from url
driver2.get("https://simple.wikipedia.org/wiki/List_of_countries")
```

```python
print(driver2.current_url)
```

```python
time.sleep(2)   # allow page to load
```

```python
# Get HTML
html2 = driver2.page_source
driver2.quit()
```

```python
# Parse HTML.
soup2 = BeautifulSoup(html2, "html.parser")
```

```python
# Locate the main content area through inspect and paste it here
content = soup2.find("div", {"class": "mw-parser-output"})
```

```python
# Extract all <a> tags inside the content
links = content.find_all("a")
```

```python
# Build a raw list of link texts
raw_list = [link.get_text(strip=True) for link in links]
```

```python
# Filter the list to keep only real country names
countries = [item for item in raw_list 
             if item.isalpha() and len(item) > 2]
countries = list(set(countries))  # remove duplicates
countries.sort()
```

```python
countries
```

the reuslts of scraping after cleaning show that the list include also regions names and not only 195  countries. therefore, i decided to Filter them using the pycountry.

```python
import pycountry

real_countries = []

for c in pycountry.countries:
    # ISO 3166-1 alpha_2 codes exist ONLY for sovereign states
    if hasattr(c, "alpha_2"):
        real_countries.append(c.name.lower())
```

```python
real_countries
```

```python
countries = [c.lower() for c in countries]
```

```python
countries_clean = [c for c in countries if c in real_countries]
```

```python
countries_clean
```

despite the cleaning still contain region names, but i decided to use it as it is for the lookup 1.5 task

```python
with open("Scraped List of Countries.txt", "w", encoding="utf-8") as f:
    for country in countries_clean:
        f.write(country + "\n")
```
