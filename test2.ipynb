```python
# Create random test data
data = pd.DataFrame({
    "Name": ["Anna", "Ben", "Clara", "David", "Eva"],        # Text column
    "Score": np.random.randint(50, 100, size=5),             # Numeric column
    "Category": ["A", "B", "A", "C", "B"]                    # Optional extra text column
})

data
```

```python
data['Score %']= data['Score'].apply(lambda x:x/100)
```

```python
data['Category_encoded'] = data['Category'].apply(lambda x: 1 if x=='A' else (2 if x=='B' else 3))
```
