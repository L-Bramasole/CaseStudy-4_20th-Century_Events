# Tokenization

1. Text is just a long string — tokenization turns it into usable pieces like words, sentences
2. It lets you clean the text properly like stopwords, punctations..
3. It prepares the text for statistical analysis like FreqDist

```python
# Import txt file

with open('key_events_20th_century.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

# Removing Stopwords

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in sentences

# filtered_sent = []
# for word in tokenized_sent:
#     if word not in stop_words:
#         filtered_sent.append(word)
# print("Tokenized Sentence:", tokenized_sent)
# print("Filterd Sentence:", filtered_sent)b
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

 after removing the stopword we see a lot of punctuation marks that  should be cleaned

# Removing punctation marks

```python
# Substitute all punctuations marks with a space 
import re  # Import the regular expressions module

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization second time 

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

# Removing new list of stopwords

```python
# cleaning new list of stopwords
new_stopwords = ["And", "Then", 'In', 't', 'th','sd','s','S','II', 'The']
```

```python
# filter out the new stopwords
filtered2 = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered2.append(word)
```

```python
filtered2
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(filtered2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

# POS Tags with TextBlob

Part‑of‑Speech (POS) Tagging
can label each word with its grammatical role. it takes the text as normal string and not tokenized. therefore if the text is already tokenized we need to clean it from " ". 

NN → noun

VB → verb

JJ → adjective

RB → adverb

PRP → pronoun

```python
#to apply TextBlob function we need to remove the tokanization
clean_text = " ".join(filtered2)
text = TextBlob(clean_text)
#TextBlob automatically splits your text into words and sentences.
# but it puts it in a container and become a TexBlob Object that contains the text string
```

```python
text
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
#converting the list to dataframe to use groupby later
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

```python
top20 = df_t.nlargest(20, 'Words')
```

```python
plt.figure(figsize = (20, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top20).set_title("number of words per tag in the 20th century key events")
```

NNP (Proper Nouns) like countries, leaders, wars, cities; NN (Common Nouns) likw war, conflict, goverment; JJ (Adjectives) are the top3 of the words.

## Nouns counts

```python
Noun_df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS")]
Noun_df.columns = ["Word", "Occurences"]
x1 = Noun_df.groupby('Word').count().reset_index()
y1 = x1.sort_values(by = ['Occurences'], ascending=False)
Noun_top15 = y1.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = Noun_top15).set_title("Nouns occurences in the key_events of 20th_century")
```

```python
NounP_df = df_text[(df_text['Word type'] == "NNP")]
NounP_df.columns = ["Word", "Occurences"]
x2 = NounP_df.groupby('Word').count().reset_index()
y2 = x2.sort_values(by = ['Occurences'], ascending=False)
NounP_top15 = y2.nlargest(15, 'Occurences')
```

# Look up for Countries and find the frequency 

```python
# Step 1: Create a list of countries
countries = [
    "Germany", "France", "Japan", "United States", "Russia", "China",
    "United Kingdom", "Italy", "Canada", "Australia"
]
```

```python
# Step 2: Convert list to DataFrame
countries_df = pd.DataFrame(countries, columns=["Country"])
countries_df
```

```python
# Step 4: Count frequency of each country in the string cleaned_text from before
countries_df["Frequency"] = countries_df["Country"].apply(
    lambda x: clean_text.count(x)
)

countries_df
```

```python
# Step 5: Sort by frequency
countries_sorted = countries_df.sort_values(by="Frequency", ascending=False)
countries_sorted
```

```python
plt.figure(figsize=(20, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(
        x="Country",
        y="Frequency",
        data=countries_sorted, 
        saturation=0.9
    ).set_title("Country Occurancies in 20th Century Key Events")
```

Insights

Japan is mentioned the most because the key events in 20th century focuses heavily on World War II, especially events in the Pacific (e.g., Pearl Harbor, Japanese expansion, atomic bombings).

Germany is next because it is central to both World Wars.

The United States, France, Russia, Italy, China appear often but play more distributed roles across different events.

Australia, the UK, and Canada appear less because they are involved mainly as allies rather than primary actors.
